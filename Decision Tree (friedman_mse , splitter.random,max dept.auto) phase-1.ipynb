```python
# Create python libraries 
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

```python
# To Read the excel sheet by using dataset 
dataset=pd.read_csv('50_Startups.csv')
```

```python
dataset
```

```python
# To convert categorical data into numerical data by using nomial (one hot encoding) Techinic
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
# To Seprate or find the input and output variable by using columns 
dataset.columns
```

```python
# To Seperate Input(independant) & Output (dependant) variables
independant=dataset[['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']]
dependant=dataset['Profit']
```

```python
independant
```

```python
dependant
```

```python
# To Split Train set value & Test set values 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independant,dependant,test_size=0.30,random_state=0)
```

```python
# To improve the r2 value by using standardisation method 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
# To find best model by using decision tree regression method 
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='best',max_features=None) 
regressor=regressor.fit(X_train,y_train)
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html 
# for more accurate model to calculate r2 value in decision tree 
```

```python
# Decision Tree Graphical method 
import matplotlib.pyplot as plt
from sklearn import tree 
tree.plot_tree(regressor)
plt.show()
```

```python
# To Predict the X test value 
y_pred=regressor.predict(X_test)
```

```python
# To Evalution of metric to find r2 value (if r2 value ~~ 1 this is best model & if r2 value ~~ 0 it is bad model)
from sklearn.metrics import r2_score 
r_score = r2_score(y_test,y_pred)
r_score
```

```python
# Save the best model by using pickle library 
import pickle
filename="finialized_model_decision_tree.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
# LOAD THE BEST MODEL FOR CALL TO ACTION 
loaded_model=pickle.load(open("finialized_model_decision_tree.sav",'rb'))
```
